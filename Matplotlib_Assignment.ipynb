Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data 
plot a scatter plot.

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.title('Scatter plot example')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

This method adds another plot to the current figure at the specified grid position.

```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots(1,1)
ax.plot([0, 1, 2, 3, 4, 5],[0, 100, 200, 300, 400, 500])
ax.plot([0, 1, 2, 3, 4, 5],[50, 20, 40, 20, 60, 70])
ax.plot([0, 1, 2, 3, 4, 5],[10, 20, 30, 40, 50, 60])
ax.plot([0, 1, 2, 3, 4, 5],[200, 350, 250, 550, 450, 150])
```
